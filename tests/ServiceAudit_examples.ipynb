## create_big_service_table

Once you find parameters you'd rather use, you can pass them to the create_big_service_table function using parameter_dict

```python
q.create_big_service_table(parameter_dict={'Outreach':{'partial_session_threshold':10,'incarceration_enrollment_pct':.7}})
```

You can also mess around with which service labels to count as enrolled by tweaking the received_service_statuses dictionary

```python
q.create_big_service_table(received_service_statuses = {
        'Legal': ['Enrolled','Aftercare'],
        'Case Management': ['Enrolled', 'Aftercare'],
        'Outreach': ['Enrolled', 'Enrolled Enough', 'Enrolled Enough (Incarcerated)', 'Aftercare'],
        'Civic Engagement': ['Enrolled'],
        'Victim Services': ['Enrolled'],
        'Therapy': ['Enrolled']})
```

```python
# the output of big_service_table can be accessed with the 'all_service_statuses' attribute
q.all_service_statuses
```
